```python
# 시퀀스 생성
xn = [0,1,2,3]
N = len(xn)
Xk = DTFS(xn, N)
print("|X(n)|=", np.abs(Xk))
ixn = IDTFS(Xk,N)
print("x(n)=", ixn)

n = np.arange(N)

# 그래프
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.stem(n, np.abs(Xk),'blue'); plt.xticks(range(0,4)); plt.grid()
plt.title("IDTFS of DTFS coefficients X(k)"); plt.ylabel("|X(k)|")
plt.subplot(2,1,2); plt.stem(n, ixn,'green'); plt.xticks(range(0,4)); plt.grid()
plt.xlabel('k'); plt.ylabel("x(n)")
```

```python
# 이산주기신화와 주기로부터 DTFS 계수 구하는 함수
def DTFS(xn, N):    
    WN = np.exp(-1j*(2*np.pi/N))                # 복소지수항
    Xk = np.zeros(N, dtype="complex64")          # DTFS 계수 어레이 설정
    for i in range(N):
        sum = 0
        for j in range(N):
            sum = sum + xn[j]*np.power(WN,i*j)   # 개별 DTFS 계수 계산
        Xk[i] = sum
    return Xk
```

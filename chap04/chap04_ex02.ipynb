## import Lib

```python
import numpy as np
import matplotlib.pylab as plt
```

## code 

```python
# 시퀀스 생성
N = 500                                         # Omega는 연속 값이지만, 연산량을 줄이기 위해서 꽤 높은 수인 500을 N으로 설정
n = np.arange(30)                               # n: 순서시퀀스 
xn = np.power(0.5, n)                           # x(n): 이산시간신호(0부터 발생, u(n) 곱해준 효과)
Omega = np.arange(-2*N, 2*N)*2*np.pi/N         # Omega: 디지털 라디안주파수 (w/fc) , len(Omega)=2000
X = np.exp(1j*Omega)/(np.exp(1j*Omega)-0.5)    # X(Omega): DTFT 변환 신호 
magX = np.abs(X)                                # magX = |X(K)|

print(len(Omega))

# 그래프 
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.stem(n,xn,'b'); plt.grid()
plt.title("x(n) & X(Omera), DTFT of x(n)"); plt.ylabel("x(n)")
plt.subplot(2,1,2); plt.plot(Omega, magX,"blue"); plt.grid()
plt.xlabel("Omega, frequency in radians"); plt.ylabel("|X(Omega)|")
plt.xlim(-4*np.pi, 4*np.pi)
```

DTFT의 두 가지 특성을 보이기 위한 실습
1. 주기성(Periodicity)
2. 대칭성(Symmetry)

## import Lib

```python
import numpy as np
import matplotlib.pylab as plt
import nbimporter
from chap04_module import *
```

## code

```python
# 시퀀스 발생
N = 500
n = np.arange(-4,7)                # 순서 시퀀스
xn = [0,0,0,1,2,3,4,5,0,0,0]       # x(n): 이산시간신호

Omega = np.arange(-2*N, 2*N)*2*np.pi/N                      # Omega: 범위 [-pi,pi]
X = np.exp(1j*Omega) + 2 + 3*np.exp(-1j*Omega) +    \       # X(Omega)
    4*np.exp(-1j*2*Omega) + 5*np.exp(-1j*3*Omega)
magX = np.abs(X)                                             # magX: |X(k)|

Omega1 = np.arange(0,N)*2*np.pi/N                            # Omega1: 범위 [0,pi]
X1 = np.exp(1j*Omega1) + 2 + 3*np.exp(-1j*Omega1) +  \      
     4*np.exp(-1j*2*Omega1) + 5*np.exp(-1j*3*Omega1)
magX1 = np.abs(X1)


# 그래프
plt.figure(figsize=(9,9))
plt.subplot(3,1,1); plt.stem(n,xn,'b'); plt.grid()
plt.title("x(n) & X(Omega), DTFT of x(n)"); plt.ylabel("x(n)"); plt.xticks(range(-4,7))
plt.subplot(3,1,2); plt.plot(Omega,magX,'b'); plt.grid()
plt.xlim(-4*np.pi, 4*np.pi); plt.ylabel("|X(Omega)|")
plt.subplot(3,1,3); plt.plot(Omega1,magX1,'b'); plt.grid()
plt.xlim(0, np.pi); plt.xlabel("Omega"); plt.ylabel("|X(Omega1)|")
```

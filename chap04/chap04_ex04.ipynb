## code

### Magnitude and pHase of X(Omega)

```python
# 시퀀스 생성
xn = [1,1,1,1]
N = len(xn)
print('x(n)=', xn)

nT = 1000
MXe = DTFT(xn, N, nT)
magMXe = np.abs(MXe)
phaMXe = np.angle(MXe, deg=True)

w = np.linspace(0, 2*np.pi, nT)


# 그래프
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.plot(w, magMXe,'b'); plt.grid()
plt.title("Magnitude and Phase of X(Omega)"); plt.ylabel("Magnitude of X(Omega)")
plt.subplot(2,1,2); plt.plot(w, phaMXe,'g'); plt.grid()
plt.xlabel("frequency in radians"); plt.ylabel("Phase of X(Omega)")

```

### Magnitude and Phase of X(k)

```python
# 시퀀스 생성
Xk = DFT(xn,N)
print("Xk=",Xk) 

magXk = np.abs(Xk)
print("Magnitude of Xk", magXk)

phaXk = np.angle(Xk, deg=True)
print("Pahse of Xk=", phaXk)

k = np.arange(N)

# 그래프
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.stem(k, magXk,'b'); plt.grid()
plt.title("Magnitude and Phase of X(k)"); plt.ylabel("Magnitude of X(k)")
plt.subplot(2,1,2); plt.stem(k, phaXk,'g'); plt.grid()
plt.xlabel("k"); plt.ylabel("Phase of X(k)")
```

### DTFT X(Omega) and DFT X(k)

```python
# 시퀀스 생성
w = np.linspace(0, np.pi, nT)
nk1 = np.arange(5)

X1 = np.zeros(5)
X1[0:4] = magXk

X2 = np.zeros(5)
X2[0:4] = phaXk

# 그래프 
# 그래프
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); 
plt.plot(w*4/3.14, magMXe, 'b'); plt.stem(nk1, X1,"magenta"); plt.grid()
plt.title("DTFT X(Omega) and DTFT X(k)"); plt.ylabel("Magnitude"); plt.xticks(range(0,5))
plt.subplot(2,1,2);
plt.plot(w*4/3.14, phaMXe, 'b'); plt.stem(nk1, X2,"magenta"); plt.grid()
plt.xlabel("k"); plt.ylabel("Phase"); plt.xticks(range(0,5))

```

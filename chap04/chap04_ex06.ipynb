고밀도스펙트럼 vs 고해상도 스팩트럼

## import Lib

```python
import numpy as np
import matplotlib.pylab as plt
import nbimporter
from chap04_module import *
```

## code

### x(n)과 X(k) 나타내기

```python
# 시퀀스 생성 
N = 10

xn = np.zeros(N, dtype = "float")
for i in range(N):
    xn[i] = np.cos(0.48*np.pi*i) + np.cos(0.52*np.pi*i)
print("x(n)=\n", xn)

Xk = DFT(xn,N)
magXk = np.abs(Xk)
print("Magnitude of X(k)=\n",magXk)

n = np.arange(N)
fn = np.linspace(0, np.pi, int(N/2)+1)
print("fn=\n",fn)

#그래프 
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.stem(n,xn,'b'); plt.grid()
plt.title("Sequence x(n) and DFT X(k), N=10"); plt.ylabel("x(n)")
plt.subplot(2,1,2); plt.stem(fn,magXk[0:int(N/2+1)],'g'); plt.grid()
plt.ylabel("|X(k)|"); plt.xlabel("frequency in radian")
```

### a. x(n)을 0~10에 대한 DFT를 결정하고 스펙트럼 나타내기 

고밀도 스펙트럼 0-10은 1, 나머지 11-100(90개)는 0으로 채운 후 DFT 수행 

```python
# 고밀도스펙트럼을 위한 스퀀스 생성
## x(n)이 1-10에 대해서만 cos(0.48*pi*n)+cos(0.52*pi*n)값을 가진다(나머지는 모두 0)
N1 = 100
xn1 = np.zeros(N1, dtype="float")
xn1[0:N] = xn
Xk1 = DFT(xn1, N1)
magXk1 = np.abs(Xk1)

n1 = np.arange(N1)
fn = np.linspace(0, np.pi, int(N1/2)+1)

# 그래프
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.stem(n1,xn1,'b'); plt.grid()
plt.title("Sequence x(n) and DFT X(k), N=100, 90 zero-padding"); plt.ylabel("x(n)")
plt.subplot(2,1,2); plt.stem(fn,magXk1[0:int(N1/2+1)],'g'); plt.grid()
plt.ylabel("|X(k)|"); plt.xlabel("frequency in radian"); plt.xticks(np.arange(0, 3.2 ,0.2))
```

### b. x(n)을 0~100에 대한 DFT를 결정하고 스펙트럼 나타내기  

고밀도스펙트럼은 처음부터 실제 데이터만으로 N=100개의 샘플을 취한 후 DFT 실행

```python
# 고해상도스펙트럼을 위한 시퀀스 생성
## x(n)이 모든 n에 대해서 cos(0.48*pi*n)+cos(0.52*pi*n)값을 가진다
N2 =100
xn = np.arange(N2, dtype="float")
for i in range(N2):
    xn[i] = np.cos(0.48*np.pi*i) + np.cos(0.52*np.pi*i)
Xk2 = DFT(xn,N2)
len(Xk2)
magXk2 = np.abs(Xk2)
print("|X(k)|=\n",magXk2)

n2= np.arange(N2)
fn = np.linspace(0, np.pi, int(N2/2)+1)

# 그래프
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.stem(n2,xn,'b'); plt.grid()
plt.title("Sequence x(n) and DFT X(k), N=100, no zero-padding"); plt.ylabel("x(n)")
plt.subplot(2,1,2); plt.stem(fn,magXk2[0:int(N2/2+1)],'g'); plt.grid()
plt.ylabel("|X(k)|"); plt.xlabel("frequency in radian"); plt.xticks(np.arange(0, 3.2 ,0.2))
```

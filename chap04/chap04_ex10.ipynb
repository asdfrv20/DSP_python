## code

```python
# 시퀀스 생성 
N = 128
n = np.arange(N)
n1 = n[0:int(N/2)]
xn = np.cos(0.48*np.pi*n) + np.cos(0.52*np.pi*n)

Xk = np.fft.fft(xn,N)
magXk = np.abs(Xk)
magXk1 = magXk[0:int(N/2)]
xnt = np.fft.ifft(Xk)
```

```python
# 그래프
plt.figure(figsize=(9,9))
plt.subplot(3,1,1); plt.stem(n,xn,'b'); plt.grid()
plt.title("Input sequence x(n), FFT X(k), IFFT xr(n)"); plt.ylabel("Input x(n)")
plt.subplot(3,1,2); plt.stem(n1,magXk1,'g'); plt.grid()
plt.ylabel("|X(k)|")
plt.subplot(3,1,3); plt.stem(n,xnt,'g'); plt.grid()
plt.ylabel("Reconstructed xr(n)"); plt.xlabel("k")
```

```python
import numpy as np
import matplotlib.pylab as plt
import nbimporter
from chap03_module import *
```

```python
# 시퀀스 발생
n = np.arange(-10,11)
alpha = -0.1+0.3*1j
xn = np.exp(alpha*n)

# 그래프
plt.figure(figsize=(9,12))
plt.subplot(4,1,1); plt.stem(n,xn,'b'); plt.grid()
plt.title('Four patrs of sequence x(n)'); plt.ylabel('Real part')
plt.subplot(4,1,2); plt.stem(n,np.imag(xn),'b'); plt.grid()
plt.ylabel('Imaginary part')
plt.subplot(4,1,3); plt.stem(n,np.abs(xn),'b'); plt.grid()
plt.ylabel('Maginitude part')
plt.subplot(4,1,4); plt.stem(n,np.angle(xn)*180/np.pi,'b'); plt.grid()
plt.ylabel('Phase part'); plt.xlabel('n')
```

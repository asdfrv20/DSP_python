```python
# 시퀀스 발생
xn = [3,11,7,0,-1,4,2]
n1 = [-3,-2,-1,0,1,2,3]
hn = [2,3,0,-5,2,1]
n2 = [-1,0,1,2,3,4,]
fn, fhn = sigfold(n2,hn)
```

```python
# 방법3: h(k)을 반전-이동시킨 h(n-k)를 사용하여 그래픽하게 구하는 방법
nmin = min(n1)+min(n2)
nmax = max(n1)+max(n2)
n = np.arange(nmin, nmax+1)
N = len(n)
yn = np.zeros(N)

# Convolution 
for i in range(nmin,nmax+1):
    sum = 0
    for j in range(int(min(fn)), int(max(fn)+1)):
        if ((i+j)>=min(n1) and (i+j)<=max(n1)):
            sum  = sum + xn[i+j+abs(min(n1))]*fhn[j+int(abs(min(fn)))]
    yn[i+abs(nmin)] = sum


# 결과 출력
print("x(n)=",xn)
print("n1=",n1)
print("h(n)=",hn)
print("n2=",n2)
print("y(n)=",yn)
print("n=",n)
```

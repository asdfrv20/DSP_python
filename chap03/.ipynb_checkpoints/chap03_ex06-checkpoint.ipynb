## import Lib

```python
import numpy as np
import matplotlib.pylab as plt
```

## code

```python
# 시퀀스 발생
xn = [3,11,7,0,-1,4,1]
n1 = [-3,-2,-1,0,1,2,3]
hn = [2,3,0,-5,2,1]
n2 = [-1,0,1,2,3,4]
```

```python
# 방법2: k에 따라서 x(k)와 h(n-k)의 곱의 합으로 구하는 방법 

nmin = min(n1) + min(n2)        # 컨볼루션 연산의 시작 순서
nmax = max(n1) + max(n2)        # 컨볼류션 연산의 끝 순서 

n = np.arange(nmin, nmax+1)     # 출력 y(n)의 순서시퀀스
N = len(n)                      # 출력 시퀀스의 길이
yn = np.zeros(N)

for i in range(nmin, nmax+1):   # n의 시작부터 끝까지
    sum=0
    for j in range(min(n1), max(n1)+1):                            # k의 시작부터 끝까지
        if ((i-j)>=min(n2) and (i-j)<=max(n2)):                   # (n-k)가 h(n)의 순서 내이면
            sum = sum +xn[j+abs(min(n1))] * hn[i-j+abs(min(n2))]   # 컨볼루션 연산
    yn[i+abs(nmin)] = sum       # 출력 저장 
    
print('x(n)=',xn)
print('h(n)=',hn)
print('y(n)=',yn)
```

```python
# 그래프 나타내기 전 전처리
aa = abs(nmin)
n1 = np.array(n1) + aa
n2 = np.array(n2) + aa

xg = np.zeros(N) 
hg = np.zeros(N)

xg[min(n1):max(n1)+1] = xn
hg[min(n2):max(n2)+1] = hn

# 그래프
plt.figure(figsize=(9,9))
plt.subplot(3,1,1); plt.stem(n,xg,'b'); plt.grid()
plt.ylabel('x(n)')
plt.subplot(3,1,2); plt.stem(n,hg,'b'); plt.grid()
plt.ylabel('h(n)')
plt.subplot(3,1,3); plt.stem(n,yn,'b'); plt.grid()
plt.ylabel('y(n)')
```

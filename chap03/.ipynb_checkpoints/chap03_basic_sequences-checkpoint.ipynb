````python
# 임의의 이산시간신호 (import test)
n = [-3,-2,-1,0,1,2,3,4]
xn = [2,1,-1, 0,1,4,3,7]

plt.stem(n,xn,"blue"); plt.grid()
plt.xlabel("n"); plt.ylabel("x(n)")
plt.title("Representation of a sequence x(n)")
````

## 이산신호의 종류 

### 단위샘플(unit sample) 시퀀스 or 단위임펄스(unit impulse)

````python
# 단위샘플 시퀀스 설정
n0=0; n1=0; n2=20             # 
N=n2-n1+1; n = np.arange(N)
xn = impseq(n0,n1,n2)
print("Unit sample sequence=",xn)

# 그래프
plt.stem(n,xn,"b"); plt.grid()
plt.xlabel("n"); plt.ylabel("x(n)")
plt.title("Representation of a sequence x(n)")
````

### 단위계단 시퀀스(unit step)

````python
# unit step 시퀀스 설정
n0=2; n1=0; n2=20    # n0: 단위계단시작위치, n1: 순서시퀀스 시작, n2: 끝
N = n2-n1+1          # N: 시퀀스 길이
n = np.arange(0,N)   # n: 순서 시퀀스
xn = stepseq(n0,n1,n2)
print('Unit step sequence=',xn)

# 그래프 
plt.stem(n,xn,'b'); plt.grid()
plt.xlabel('n'); plt.ylabel('x(n)')
plt.title("Unit step sequence")
````

### 실수지수(real-valued exponenetial) 시퀀스

````python
# 실수지수 함수는 numpy에서 지원하는 np.power() 함수를 활용하여 구현한다.

# 실수지수 시퀀스 설정
a = 0.7              # a: 지수 밑
N = 20               # N: 시퀀스 길이
n = np.arange(N)     # n: 순서 시퀀스
xn = np.power(a,n)   # xn: 실수지수 시퀀스
print('Real-valued exponential=',xn)

# 그래프 
plt.stem(n,xn,'b'); plt.grid()
plt.xlabel('n'); plt.ylabel('x(n)')
plt.title("Real-valued exponential sequence")
````

### 복소지수(complex-valued exponential sequence) 시퀀스

np.linspace(start,stop,num)

````python
# 복소지수 시퀀스 설정
sigma=2; freq=3                  # 복수지수 식에 들어갈 sigma와 frequncy 값
n = np.linspace(0,10,100)        # 순서시퀀스(0~10, 100등분)
xn = np.exp((sigma+freq*1j)*n)   # 복소지수 시퀀스 (amplitude)
magxn = abs(xn)                  # 복소지수의 크기를 나타내는 '절대값 복소지수 시퀀스'(magnitude) 

# 그래프
plt.figure(figsize=(8,8))
plt.subplot(2,1,1); plt.stem(n,xn,'b'); plt.grid()    # 복소지수 시퀀스
plt.ylabel('x(n), Amplitude')
plt.title('Complex-valued exponential sequence')
plt.subplot(2,1,2); plt.stem(n,magxn,'b');plt.grid()    # 절대값 복소지수 시퀀스
plt.ylabel('|x(n)|, Magnitude')
````

### 정현파(sinusoidal) 시퀀스

````python
# 정현파 시퀀스 설정
n = np.linspace(0, 10*np.pi, 100)       # n: 순서 시퀀스(0~10pi, 100등분)
xn = 3*np.cos(0.1*np.pi*n + np.pi/2)    # xn: 정편파 시퀀스(진폭(A)=3, 라디안 주파수(w)=0.1*pi, 위상=pi/2)

# 그래프
plt.figure(figsize=(8,6))
plt.stem(n,xn,'b'); plt.grid()
plt.xlabel('n[sample]'); plt.ylabel('Amplitude')
plt.title('Sinusoidal sequence')
````

### 무작위(random) 시퀀스 

````python
# random seqence
n = np.arange(20)
xn = np.random.rand(20)
print('Random sequence x(n)=',xn)

# 그래프
plt.figure(figsize=(8,6))
plt.stem(n,xn,'b'); plt.grid()
plt.xlabel('n[sample]'); plt.ylabel('Amplitude')
plt.title('Random sequence')
````

### 주기(periodic) 시퀀스

````python
# 주기 시퀀스 설정
n = np.arange(35)    # n: 순서시퀀스
x = list(range(1,8))
print('x=', x)
xn = 5*x             # xn: 주기 시퀀스
print('xn=', xn)

# 그래프
plt.stem(n,xn,'b'); plt.grid()
plt.xlabel('n[sample]'); plt.ylabel('Amplitude')
plt.title('Peoriodic sequence')
````

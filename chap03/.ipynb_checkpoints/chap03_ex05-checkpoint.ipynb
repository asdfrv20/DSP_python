## import Lib

```python
import numpy as np
import matplotlib.pylab as plt
```

## code 

```python
# 시퀀스 발생
n = np.arange(-2,5)
N = len(n)

xn = np.zeros(N)    # x(n)
xn[1]=1; xn[2]=1; xn[3]=0.5

hn = np.zeros(N)    # h(n)
hn[2]=1; hn[3]=0.5; hn[4]=0.25

yn = np.zeros(N)    # y(n)

# 방법1: x(n)의 각각의 입력에 대한 임펄스응답의 중첩 합을 구하는 방법 적용하여 연산
for i in range(N):
    if xn[i] != 0:
        print("n=", i)
        for k in range(3):
            yn[i+k] = yn[i+k] + xn[i]*hn[2+k]
        print("y(n)=",yn)
    
# 그래프로 나타내기 
plt.figure(figsize=(9,6))
plt.subplot(3,1,1); plt.stem(n,xn,'b')
plt.subplot(3,1,2); plt.stem(n,hn,'b')
plt.subplot(3,1,3); plt.stem(n,yn,'b')
```

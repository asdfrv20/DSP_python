## 문제 a

```python
# 시퀀스 발생 시키기 
n = np.arange(-5,6)
xn = 2*impseq(-2,-5,5) - impseq(4,-5,5)

# 그래프 
plt.figure(figsize = (9,3))
plt.stem(n,xn,'b'); plt.grid()
plt.ylabel("a: x(n)")
```

## 문제 b

```python
# 시퀀스 발생
n = np.arange(0,21)
xn1 = n*(stepseq(0,0,20)-stepseq(10,0,20))
xn2 = 10*np.exp(-0.3*(n-10))*(stepseq(10,0,20)-stepseq(20,0,20))
xn = xn1 + xn2

# 그래프
plt.figure(figsize = (9,3))
plt.stem(n,xn,'b'); plt.grid()
plt.ylabel("b: x(n)")
```

## 문제 c

```python
# 시퀀스 발생
n = np.arange(0,21)
xn = np.cos(0.04*np.pi*n) + 0.2*np.random.randn(len(n))

# 그래프
plt.figure(figsize = (9,3))
plt.stem(n,xn,'b'); plt.grid()
plt.ylabel("c: x(n)")
```

## 문제 d

```python
# 시퀀스 발생
n = np.arange(-10,10)
x = [5,4,3,2,1]
xn = x*4

# 그래프
plt.figure(figsize = (9,3))
plt.stem(n,xn,'b'); plt.grid()
plt.ylabel("d: x(n)")
```

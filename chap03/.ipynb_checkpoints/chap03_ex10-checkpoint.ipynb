```python
import numpy as np
import matplotlib.pylab as plt
from scipy import signal

import nbimporter 
from chap03_module import *
```

## code

```python
# 입력계수(b)와 출력계수(a) 설정
b = [1]           # 입력계수
a = [1, -0.9]     # 출력계수

# 시퀀스 발생
n = np.arange(0,51)                         # 순서시퀀스
xn = stepseq(0,0,50) - stepseq(10,0,50)     # 입력시퀀스
hn = np.power(0.9,n)                        # 임펄스 응답시퀀스, 굳이 stepseq(0,0,50)을 곱하지 않아도 범위 제한으로 해결
yn = signal.lfilter(b,a,xn)                 # 출력시퀀스

# 그래프 
plt.figure(figsize=(9,9))
plt.subplot(3,1,1); plt.stem(n,xn,'b'); plt.grid()
plt.title("Input x(n), Impulse response h(n) and Output y(n)"); plt.ylabel("x(n)")
plt.subplot(3,1,2); plt.stem(n,hn,'b'); plt.grid()
plt.title("Input h(n), Impulse response h(n) and Output y(n)"); plt.ylabel("h(n)")
plt.subplot(3,1,3); plt.stem(n,yn,'b'); plt.grid()
plt.title("Input y(n), Impulse response h(n) and Output y(n)"); plt.ylabel("y(n)")
```

## import Lib

```python
import numpy as np
import matplotlib.pylab as plt
import nbimporter
from chap03_module import *
```

## x(n) 발생

```python
# 시퀀스 발생
n = np.arange(-10,11)
u1 = stepseq(0,-10,10)
u2 = stepseq(10,-10,10)
xn = u1 - u2

# 그래프
plt.figure(figsize = (9,3))
plt.stem(n,xn,'b'); plt.grid()
plt.ylabel('x(n)'); plt.xlabel('n')
```

## 우함수/기함수 성분 분해

`chap03_module`의 `evenodd()`를 활용한다

```python
# 성분 분해 
n, xe, xo = evenodd(n,xn)

# 그래프 
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.stem(n,xe,'b'); plt.grid()
plt.title('Even-Odd decomposition of Sequence x(n)'); plt.ylabel('xe(n)')
plt.subplot(2,1,2); plt.stem(n,xo,'b'); plt.grid()
plt.ylabel('xe(n)'); plt.xlabel('n')
```

```python
import numpy as np
```

```python
def impseq(n0, n1, n2):
    N = n2 - n1 + 1
    n = np.arange(N)
    xn = np.zeros(N)
    for i in range(N):
        if (i+n1) == n0: xn[i]=1
    return xn
```

```python
def stempseq(n0, n1, n2):
    N = n2-n1+1
    n = np.arange(N)
    xn=np.zeros(N)
    for i in range(N):
        if (i+n1)>=n0: xn[i]=1
    return xn
```

## 문제 a

n = 0, ... ,100에서 임펄스 응답 h(n)을 계산하고 그래프를 나타내어라

```python
# 시퀀스 발생
n = np.arange(101)              # 순서시퀀스
xn1 = impseq(0,0,100)            # 단위임펄스 입력시퀀스
hn = signal.lfilter(b,a,xn1)    # 임펄스응답 시퀀스

# 그래프 
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.stem(n,xn1,'b'); plt.grid()
plt.title("Impulse input x(n) and Impulse response h(n)"); plt.ylabel("x(n)")
plt.subplot(2,1,2); plt.stem(n,hn,'b'); plt.grid()
plt.xlabel("n"); plt.ylabel("h(n)")
```

## 문제 b

n=0, ... ,100에서 단위계단 응답 s(n)을 계산하고 그래프로 나타내어라 

```python
# 시퀀스 발생
xn2 = stepseq(0,0,100)          # 단위임펄스 입력시퀀스
sn = signal.lfilter(b,a,xn2)    # 단위스텝 응답 시퀀스

# 그래프 
plt.figure(figsize=(9,6))
plt.subplot(2,1,1); plt.stem(n,xn2,'b'); plt.grid()
plt.title("Unit step input x(n) and Unit step response s(n)"); plt.ylabel("x(n)")
plt.subplot(2,1,2); plt.stem(n,sn,'b'); plt.grid()
plt.xlabel("n"); plt.ylabel("s(n)")
```

## 문제 c

임펄스응담 h(n)으로 특정된 이 시스템이 안정한지를 판단하시오

```python
xn1_sum = np.sum(xn1)
hn_sum = np.sum(hn)
sum1 = hn_sum - xn1_sum
print("sum_of_unit impulse response=", sum1)

xn2_sum = np.sum(xn2)
sn_sum = np.sum(sn)
sum2 = sn_sum - xn2_sum
print("sum_of_unit step response=", sum2)
```

## 18번

h(n) = [2,-3,-1,4,-5,7,-5,4,-1,-3,2]

n = [0~10]

제2형 선형위상 FIR 필터의 크기응답, 위상응답, 영점-극점 다이어그램을 구하여라

But 문제에서 주어진 h(n)은 제1형 FIR필터...?(M=11, 대칭 형태)

### 제2형 선형위상 FIR 필터 적용한 경우

```python
# 주어진 임펄스 함수(h(n)) 설정 및 시퀀스 생성
hn=np.array([2,-3,-1,4,-5,7,-5,4,-1,-3,2])  #임펄스응답
M=len(hn)             # 필터길이
n=np.arange(0,M)      # 시퀀스축(첫 자리 0부터 시작)
Omega=np.linspace(0, np.pi, 1000)     #0에서 pi까지 1000등분
```

```python
# 문제대로 제2형으로 구한 경우
# 주파수 응답, 크기응답, 위상응답 구하기
H=np.arange(1000)
H=np.float64(H)     #주파수응답

for x in range(1000):
    H[x]=2*hn[5]*np.cos(Omega[x]*(1-1/2))+2*hn[4]*np.cos(Omega[x]*(2-1/2))\
    +2*hn[3]*np.cos(Omega[x]*(3-1/2))+2*hn[2]*np.cos(Omega[x]*(4-1/2))\
    +2*hn[1]*np.cos(Omega[x]*(5-1/2))+2*hn[0]*np.cos(Omega[x]*(6-1/2))

amplitude=H     #진폭응답
magnitude=abs(H)     #크기응답
phase=np.angle(H)*180/np.pi    #위상응답
```

```python
# 임펄스 응답
plt.figure(figsize=(8,4))
plt.stem(n,hn,"b"); plt.grid()
plt.title("Impulse Response")
```

```python
# 크기응답
plt.figure(figsize=(8,4))
plt.plot(Omega/np.pi,magnitude,"g"); plt.grid()
plt.ylabel("Magnitude"); plt.title("Magnitude Response")
```

```python
# 위상응답
plt.plot(Omega/np.pi,phase,"b"); plt.grid()
plt.ylabel("Phase"); plt.xlabel("Frequency in pi radians")
plt.title("Phase Response")
```

```python
# 극점-영점 다이어 그램
plt.figure(figsize=(5,5))
an=np.array([1,0]); zplane(hn,an); plt.grid()
plt.title("Pole-Zero Diagram")
```

### 제 1형 FIR 필터로 계산한 경우

```python
# 제1형 FIR 필터로 계산한 경우
H=np.arange(1000); H=np.float64(H)     #주파수응답
for x in range(1000):
    H[x]=hn[5]*np.cos(Omega[x]*0)+2*hn[4]*np.cos(Omega[x]*1)\
    +2*hn[3]*np.cos(Omega[x]*2)+2*hn[2]*np.cos(Omega[x]*3)\
    +2*hn[1]*np.cos(Omega[x]*4)+2*hn[0]*np.cos(Omega[x]*5)
amplitude=H     #진폭응답
magnitude=abs(H)     #크기응답
phase=np.angle(H)*180/np.pi    #위상응답
```

```python
# 임펄스 응답
plt.figure(figsize=(8,4))
plt.stem(n,hn,"b"); plt.grid()
plt.title("Impulse Response")
```

```python
# 크기응답
plt.figure(figsize=(8,4))
plt.plot(Omega/np.pi,magnitude,"g"); plt.grid()
plt.ylabel("Magnitude"); plt.title("Magnitude Response")
```

```python
# 위상응답
plt.plot(Omega/np.pi,phase,"b"); plt.grid()
plt.ylabel("Phase"); plt.xlabel("Frequency in pi radians")
plt.title("Phase Response")
```

```python
# 극점-영점 다이어 그램
plt.figure(figsize=(5,5))
an=np.array([1,0]); zplane(hn,an); plt.grid()
plt.title("Pole-Zero Diagram")
```

## 19번

"카이저 창"을 이용한 '선형위상 LPF' 설계 및 실현

통과대역 차단주파수: 0.2pi, 저지대역 차단주파수: 0.3pi, 저지대역 감쇄 값: As=50dB인 

```python
# 각 파라미터 설정 및 시퀀스 생성
## 설정 요소: 창 길이(M), 저지대역 감쇄(As), 베타(beta), 시퀀스 축
## As=50bB이기 때문에 Hamming 창에 근사하다고 생각하여,
## M=66, beta=5 로 설정하였습니다.(교재 표7-3,표7-4 참고)
M=66       #창 길이
As=50      #저지대역 감쇄
beta=5     #베타
nt=np.arange(0,M)     #시퀀스 축
```

```python
# 임펄스 함수(hd) 및 창 함수(wn) 생성 및 곱하기
Omega_s=0.2*np.pi              # 저지대역 차단주파수
Omega_p=0.3*np.pi              # 통과대역 차단주파수
Omega_c=(Omega_p+Omega_s)/2    # 차단주파수
print(Omega_c)                 # 차단주파수 확인

hd=ideal_lp(np.pi,M)-ideal_lp(Omega_c,M)  # 이상적인 임펄스응답
wn=signal.kaiser(M,beta)     #카이저 창 함수
hn=hd*wn     #유한 임펄스응답
```

```python
# 주파수 응답 구하기
nf,H=signal.freqz(hn)     #주파수축, 주파수응답
H_dB=20*np.log10(abs(H))     #주파수응답(dB)
```

```python
# 임펄스 응답
plt.figure(figsize=(9,3))
plt.stem(nt,hn,"b"); plt.grid()
plt.ylabel("h(n)"); plt.xlabel("n"); plt.xlim(0,M-1)
plt.title("Impulse response, Frequency response, Frequency filtering result")
```

```python
# 크기응답
plt.figure(figsize=(9,3))
plt.plot(nf/np.pi,H_dB,"b"); plt.grid(); plt.xlim(0,1)
plt.xlabel("Frequency in pi radians"); plt.ylabel("Magnitude [dB]")
```

```python
# 처프신호에 대한 주파수 선별 결과
plt.figure(figsize=(9,3))
t=np.linspace(0,1,5000)
xn=chirp(t,f0=10, t1=0.2, f1=500, method="linear") #처프신호 생성
yn=np.convolve(hn,xn) #컨볼루션
plt.plot(yn,"b"); plt.xlim(0,5000)
plt.xlabel("Samples(Frequency(0~5000[Hz]))"); plt.grid()
```

## 20번

"Hamming 창-설계 기법"을 이용한 '선형위상 BSF' 설계 및 실현

1. 하위 및 상위 저지대역 에지주파수: 0.4pi, 0.6pi, As=50[dB]
2. 하위 및 상위 통과대역 에지주파수: 0.3pi, 0.7pi, Rp=0.25[dB]

```python
# 각 파라미터 설정
Omega_p1=0.4*np.pi; Omega_p2=0.6*np.pi   # 2개의 Omega_p
Omega_s1=0.3*np.pi; Omega_s2=0.7*np.pi   # 2개의 Omega_s
tr_width=np.min([(Omega_p1-Omega_s1),(Omega_s2-Omega_p2)])     #천이대역폭
M=int(np.ceil(6.6*np.pi/tr_width))+1; print("M=",M)     #창 길이
Omega_c1=(Omega_s1+Omega_p1)/2; Omega_c2=(Omega_s2+Omega_p2)/2     #차단주파수
nt=np.arange(0,M)     #시퀀스 축
```

```python
# 임펄스 응답(hd, 창 함수(wn) 구하고 곱해주기 
hd=ideal_lp(Omega_c1,M)+(ideal_lp(np.pi,M)-ideal_lp(Omega_c2,M)) #BSF 임펄스응답
wn=signal.hamming(M)     #해밍 창 함수
hn=hd*wn     #유한 임펄스응답
```

```python
# 주파수 응답 구하기 
nf,H=signal.freqz(hn)     #주파수축, 주파수응답
H_dB=20*np.log10(abs(H))     #주파수응답(dB)
```

```python
# 임펄스 응답
plt.figure(figsize=(9,3))
plt.stem(nt,hn,"b"); plt.grid()
plt.ylabel("h(n)"); plt.xlabel("n"); plt.xlim(0,M-1)
plt.title("Impulse response, Frequency response, Frequency filtering result")
```

```python
# 크기응답
plt.figure(figsize=(9,3))
plt.plot(nf/np.pi,H_dB,"b"); plt.grid(); plt.xlim(0,1)
plt.xlabel("Frequency in pi radians"); plt.ylabel("Magnitude [dB]")
```

```python
# 처프신호에 대한 주파수 선별 결과
t=np.linspace(0,1,5000)
xn=chirp(t,f0=10, t1=0.2, f1=500, method="linear") #처프신호 생성
yn=np.convolve(hn,xn) #컨볼루션으로 출력생성

plt.figure(figsize=(9,3))
plt.plot(yn,"b"); plt.xlim(0,5000)
plt.xlabel("Samples(Frequency(0~500[Hz]))"); plt.grid()
```

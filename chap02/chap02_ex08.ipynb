# chap02_ex08: 영차유지 보간(ZOH: Zero-Order hold interpolation)

## import Libarary

```python
# import Library
import numpy as np
import matplotlib.pylab as plt
from scipy.interpolate import *
```

## 신호 정의

```python
# x(n) 신호 정의
Ts = 0.0002                        # 샘플링 주기
n = np.arange(-25,26)             # 순서시퀀스 어레이
xn = np.exp(-1000*np.abs(n*Ts))   # 샘플 시퀀스
```

```python
# ZOH를 활용한 x(t) 정의
ft = interp1d(n, xn, kind="previous")    # 영차보간 적용
tn = np.linspace(-25,25,1000)
```

## 그래프 나타내기 

```python
# 그래프로 나타내기 
plt.figure(figsize=(12,9))
plt.subplot(2,1,1); plt.stem(n,xn,"blue"); plt.ylabel("x(n)"); plt.grid()
plt.title("Singal recontstruction using zero-oreder interpolation")
plt.subplot(2,1,2); plt.plot(tn,ft(tn),"red"); plt.grid()
plt.xlabel("tn"); plt.ylabel("xa(t)")
plt.stem(n,xn,"blue",".")    # 덧그리기 

```

# DSP: 예제 2-2 

```python
#chap02_ex02.py
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs=100; dt=1/fs; N=300
t = np.arange(0,N)*dt

# 주파수 1Hz(x1t), 3Hz(x3t), 19Hz(x19t) 정현파 신호들
# x와 t사이에 들어간 숫자는 곱해진 주파수의 크기를 말합니다 
x1t = 1.0*np.sin(2*np.pi*t)
x3t = 1.0*np.sin(2*np.pi*3*t)
x19t = 1.0*np.sin(2*np.pi*19*t)
```

```python
# 10개의 신호를 합성한 신호 xt

sum = 0
generate_sig_num = 0       # 발생시킨 신호의 수를 count
for i in range(1,20,2):    # 1~19까지 step=2로 발생 총 10개의 신호 
    sum = sum+1/i*np.sin(2*np.pi*i*t)
    generate_sig_num += 1 
xt = sum
print("generate_sig_num: ",generate_sig_num)
```

```python
# 주파수 영역에서 Spectrum으로 신호 xf 분석(Xf)
df = fs/N
f = np.arange(0,N)*df
Xf = np.fft.fft(xt)
```

```python
# 그래프로 나타내기 
plt.subplot(5,1,1); plt.plot(t,xt,"b"); plt.ylim(-1,1); plt.grid()
plt.ylabel("x(t)")

# plt.title("A periodic signal x(t)=x1(t)+x2(t)+...+x19(t)")
plt.subplot(5,1,2); plt.plot(t,x1t,"b"); plt.ylim(-1,1); plt.grid()
plt.ylabel("x1(t)")
plt.subplot(5,1,3); plt.plot(t,x3t,"b"); plt.ylim(-1,1); plt.grid()
plt.ylabel("x3(t)")
plt.subplot(5,1,4); plt.plot(t,x19t,"b"); plt.ylim(-1,1); plt.grid()
plt.ylabel("x19(t)")
    
plt.subplot(5,1,5); plt.plot(f[0:int(N/2+1)],np.abs(Xf[0:int(N/2+1)]),"b");
plt.xlabel("frequency(Hz)"); plt.ylabel("Spectrum X(f)"); plt.grid()

plt.show()
```

# 10개의 신호 성분들을 모두 그래프로 확인해보기 

```python
# 10개의 신호를 전부 확인해 보자 
xt_element = []
for n in range(1,11):
    temp = 1.0*np.sin(2*np.pi*(2*n-1)*t)
    xt_element.append(temp)
len(xt_element)
```

```python
# x1t ~ x19t 그래프 확인
for n in range(1,11):
    plt.subplot(10,1,n); plt.plot(t,xt_element[n-1],"b"); plt.ylim(-1,1); plt.grid();
    plt.ylabel('x'+str(n)+'(t)')
```

```python
# 5개씩 그래프 확인(1)
for n in range(1,6):
    plt.subplot(5,1,n); plt.plot(t,xt_element[n-1],"b"); plt.ylim(-1,1); plt.grid();
    plt.ylabel('x'+str(2*n-1)+'(t)')
```

```python
# 5개씩 그래프 확인(2)
for n in range(6,11):
    plt.subplot(5,1,n-5); plt.plot(t,xt_element[n-1],"b"); plt.ylim(-1,1); plt.grid();
    plt.ylabel('x'+str(2*n-1)+'(t)')
```

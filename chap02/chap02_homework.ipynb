## 연습문제 3

### (1)번 문제

```python
Question_num = 'ch02_num3_1'
```

```python
# 신호 정의
fs=100; dt=1/fs; N=100  # 주파수, 시간간격, 샘플링 수 정의 
t = np.arange(-N,N)*dt  # N*dt = 300*0.01 = 3[s]

x1t = 1.0 * np.cos(2*1*np.pi*t)              #x1t 정의
x2t = 1.0 * np.cos(2*2*np.pi*t + np.pi/3)    # x2t 정의 
xt = x1t + x2t          # xt 정의

# Spectrum
df = fs/N               # 주파수간격: df = fs/N = 100/100 = 1[Hz]
f = np.arange(0,N)*df   # 주파수축 : N*df = 100*1 = 100
Xf = np.fft.fft(xt)     # 주파수별 크기 계산
```

```python
# 그래프로 나타내기 
plt.figure(figsize=(12,16))
# x1t
plt.subplot(4,1,1); plt.plot(t,x1t,'b'); plt.grid()
plt.xlabel("t"); plt.ylabel("x1(t)")
plt.title(Question_num)
# x2t
plt.subplot(4,1,2); plt.plot(t,x2t,'b'); plt.grid()
plt.xlabel("t"); plt.ylabel("x2(t)")
#xt
plt.subplot(4,1,3); plt.plot(t,xt,'b'); plt.grid()
plt.xlabel("t"); plt.ylabel("x(t)")
# Spectrum
plt.subplot(4,1,4); plt.plot(f[0:int(N/2+1)],np.abs(Xf[0:int(N/2+1)]),'b'); plt.grid()
plt.xlabel("f"); plt.ylabel("X(f)")

plt.savefig('D:\GitHub_asdfrv20\DSP_python\chap02\chap02_img\practice' + Question_num)
```

## 연습문제 6

```python
Question_num = 'ch02_num6'
```

```python
# 기본요소 정의(fs,dt,N,t,T) 
fs=100; dt=1/fs
t=np.arange(-2,2,0.001)

# x(t): 구형파 정의 
xt = np.zeros(len(t))
for n in range(len(t)):
    if ( (t[n]>=-1) and (t[n]<=1)): xt[n] = 1    
    else: xt[n] = 0

# X(f): Fourier Transform을 통한 X(f) 연산
w = np.arange(-2*np.pi, 2*np.pi, 0.1)
N = len(w)
Xw = np.zeros(N)
T = 1
for i in range(-int(N/2),int(N/2)):
    if i == 0:
        Xw[i+int(N/2)] = 1
    else:
        Xw[i+int(N/2)] = np.sin(i*T/2)/(i/2)
```

```python
# 그래프로 나타내기
plt.figure(figsize=(12,6))

# xt
plt.subplot(2,1,1); plt.plot(t,xt,'b'); plt.grid()
plt.xlabel("t"); plt.ylabel("x(t)"); plt.xlim(-2,2);plt.title(Question_num)

# Xf
plt.subplot(2,1,2); plt.plot(w,Xw,'b'); plt.grid()
plt.xlabel("w"); plt.ylabel("X(w)")

plt.savefig('D:\GitHub_asdfrv20\DSP_python\chap02\chap02_img\practice' + Question_num)
```

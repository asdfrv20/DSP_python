# chap02_ex06: sinc 함수 그래프 그리기

```python
import numpy as np
import matplotlib.pylab as plt
```

## w, N, Xw, T 초기화 (주파수축, 주파수 축 길이, Spectrum으로 나타냈을 때의 X(w), 주기)

```python
w = np.arange(-2*np.pi, 2*np.pi, 0.1)  # 주파수축 설정
N = len(w)        # 전체 주파수축 길이
Xw = np.zeros(N)  # CTFT 결과 계수 저장 어레이
T = 1             # 주기=1
```

```python
# 확인
print('w: ', w)
print('N: ', N)
print('Xw: ', Xw)
print('T: ', T)
```

## CTFT의 계수 X(f) 계산

```python
for i in range(-int(N/2),int(N/2)):
    if i==0:
        Xw[i+int(N/2)] = 1
    if i!=0:
        Xw[i+int(N/2)] = np.sin(i*T/2)/(i/2)    # CTFT 계수 X(f) 계산
        
```

## 그래프 그리기

```python
plt.figure(figsize=(12,9))   # 그래프 사이즈 조절 
plt.plot(w,Xw,"blue")
plt.stem(w,Xw,"blue"); plt.grid()
plt.xlabel("w, frequency in radians"); plt.ylabel("X(w)")
plt.title("CTFT, of a non-periodic signal")
```
